```python
import pandas as pd

df = pd.read_csv("student-mat.csv", sep=';')
df.head()
```

```python
df.head()

print("Missing Values in Each Column:")
print(df.isnull().sum())

print("\nData Types of Each Column:")
print(df.dtypes)

print("\nShape of the Dataset (rows, columns):")
print(df.shape)
```

```python
print("Duplicate rows before cleaning:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicate rows after cleaning:", df.duplicated().sum())

print("Final shape of the dataset:", df.shape)
```

```python
average_g3 = df["G3"].mean()
print("1. Average Final Grade (G3):", average_g3)

above_15_count = df[df["G3"] > 15].shape[0]
print("2. Students scoring above 15 in G3:", above_15_count)

correlation = df["studytime"].corr(df["G3"])
print("3. Correlation between Study Time and G3:", correlation)

gender_avg = df.groupby("sex")["G3"].mean()
print("4. Average G3 by Gender:\n", gender_avg)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
plt.hist(df['G3'], bins=10, color='orchid', edgecolor='black')
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.grid(True)
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x="studytime", y="G3", data=df)
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time (Weekly)")
plt.ylabel("Final Grade (G3)")
plt.grid(True)
plt.show()

plt.figure(figsize=(6,5))
sns.barplot(x="sex", y="G3", data=gender_avg, hue="sex", palette="pastel", legend=False)
plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade (G3)")
plt.grid(True)
plt.show()
```
